```python
import os
import sys
import time
import json
import warnings
warnings.filterwarnings(action='ignore', category=UserWarning, module='gensim')
from gensim.models import Word2Vec
import logging
```

```python
folders = [x for x in os.listdir('..//..//data//clean_data') if x[-3:] != 'csv']
```

```python
sizes = [300]
data = []
if 'model' not in os.listdir('..//..//data') : os.system('mkdir "..//..//data//model"') 
for folder in folders :
    os.system('mkdir "..//..//data//model//%s"'%folder)
    #data = []
    for file in os.listdir('..//..//data//clean_data//%s'%folder) :
        data+= json.load(open('..//..//data//clean_data//%s//%s'%(folder,file)))

#######
for dimsize in sizes :
    os.system('mkdir "..//..//data//model//%s//%s"'%(folder,dimsize))
    logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO) 
    model = Word2Vec(data, size= dimsize, window = 5 , min_count= 5 , workers=4, sg=1)
    model.save('..//..//data//model//%s//%s//model'%(folder,dimsize))
    model = ''
```
