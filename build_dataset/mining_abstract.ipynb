```python
ablist = os.listdir('../data/preprocessing/abstract_extract/')
outdir = '../data/abstract_full/'
count = 0
track = 0
save = {}
for each in ablist :
    data = json.load(open('../data/preprocessing/abstract_extract/%s'%each))
    for ID in data :
        save[ID] = data[ID]
        count+=1
        if count % 50000 == 0 :
            outfile = open(os.path.join(outdir,os.path.basename('abstract_%s.json'%track)),'w')
            outfile.write(json.dumps(save))  
            outfile.close()
            print(track)
            track+=1
            save = {}

outfile = open(os.path.join(outdir,os.path.basename('abstract_%s.json'%track)),'w')
outfile.write(json.dumps(save))  
outfile.close()
print(track)
```
