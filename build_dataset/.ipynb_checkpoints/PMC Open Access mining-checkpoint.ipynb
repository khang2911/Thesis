```python
import urllib.request
import os
import sys
import json
from bs4 import BeautifulSoup
from urllib.request import urlopen
import time
```

```python
##### Change work directory to folder tar - set up download location
os.chdir('..\\data\\tar')
```

```python
##### Take PMC id from json file 
idlist = json.load(open('..\\id.json'))
##### list contain ID which can not be dowloaded
misslist = []
##### Take downloaded file list - when there is the interuption from previous dowdload and we want to continue with same idlist
dowloaded = os.listdir('..\\tar')
dowloadedlist = []
for each in dowloaded :
    dowloadedlist.append(each.split('.')[0][3:])
```

```python
##### Down tar.gz file from PMC open access subset
miss = 0
success =0
begintime = time.time()
for ID in idlist :
    if ID not in dowloadedlist :
        try :
            filename = 'PMC' + ID + '.tar.gz'
            url = 'https://www.ncbi.nlm.nih.gov/pmc/utils/oa/oa.fcgi?id=PMC' + ID
            page = urlopen(url)
            soup = BeautifulSoup(page,'html.parser')
            link = soup.find('link').get('href')
            page.close()
            #####
            file = open(filename, 'wb')
            down = urllib.request.urlretrieve(link,filename)
            file.close()
            success+=1

        except :
            misslist.append(ID)
            miss+=1
    else :
        success+=1
    ######
    print('%s / %s / %s'%(success,miss,len(idlist)))
#####
totaltime = time.time() - begintime
print('Process takes %s seconds'%totaltime)
```
